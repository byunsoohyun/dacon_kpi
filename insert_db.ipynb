## insert customers

```python
customers
```

```python
metadata = MetaData()
```

```python
table = Table("customers", metadata,
      Column('Customer_id', String(100)),
      Column('Customer_unique_id', String(100)),
      Column('Customer_zipcode_prefix', Integer),
      Column('Customer_city', String(100)),
      Column('Customer_state', String(100)),
     )
```

```python
metadata.create_all(engine)
```

```python
customers.to_sql("customers", if_exists='append', con=engine, index=False)
```
